#### The regression analysis aims to analyze the relationship between height and weight in olympic medalist. Therefore, athletes wihout a medal will be excluded from the analysis

```python
# Create a subset excluding 'na's

medals = df.loc[df['medal'].isin(['Gold','Silver','Bronze'])]
```

```python
# Create now a data frame filtering rows with populated values in 'height' and 'weight'

medals_nn = medals[(medals['height'].notnull()) & (medals['weight'].notnull())]
```

# 04. Data prep for regression analysis

```python
# Create a scatterplot using matplotlib for another look at how the chosen variables plot against each other.

medals_nn.plot(x = 'height', y='weight',style='o') # The style option creates a scatterplot; without it, we only have lines.
plt.title('Height vs Weight of Olympic Medalists')  
plt.xlabel('Height')  
plt.ylabel('Weight')  
plt.show()
```

### There are some outliers in that are over 160 kg but I will keep them in the analysis for now. 

## Hypothesis: In Olympic Medalist, the heavier the athlete, the bigger their height is. 

```python
# Reshape the variables into NumPy arrays and put them into separate objects.

X = medals_nn['weight'].values.reshape(-1,1)
y = medals_nn['height'].values.reshape(-1,1)
```

```python
X
```

```python
y
```

```python
# Split data into a training set and a test set.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

# 05. Regression Analysis

**Run Linear Regression on Data**

```python
# First, Create a regression object.

regression = LinearRegression()  # This is the regression object, which will be fit onto the training set.
```

```python
# Fit the regression object onto the training set.

regression.fit(X_train, y_train)
```

```python
# Second, Predict the values of y using X.

y_predicted = regression.predict(X_test)
```

```python
# Create a plot that shows the regression line from the model on the test set.

plot_test = plt
plot_test.scatter(X_test, y_test, color='gray', s = 15)
plot_test.plot(X_test, y_predicted, color='red', linewidth =3)
plot_test.title('Height vs. Weight of Olympic Medalist (Test set)')
plot_test.xlabel('Height')
plot_test.ylabel('Weight')
plot_test.show()
```

The linear model represented by the red regression line doesn’t perfectly cover all the data points. In the area where the weight get above 200 kg, for example, there are still many data points that deviate from the regression line and tend towards the shorter height

### Check the model performance statistics—MSE and R2 score

```python
# Create objects that contain the model summary statistics.

rmse = mean_squared_error(y_test, y_predicted) # mean squared error
r2 = r2_score(y_test, y_predicted) # R2 score. 
```

```python
# Print the model summary statistics. This is where you evaluate the performance of the model.

print('Slope:' ,regression.coef_)
print('Mean squared error: ', rmse)
print('R2 score: ', r2)
```
