# 03 . Create variable 'Season'

```python
# Convert 'edition' column to strings

df['edition'] = df['edition'].astype(str)
```

```python
# Check change of data type

df.dtypes
```

```python
def extract_season(edition):
    if 'Summer' in edition:
        return 'Summer'
    elif 'Winter' in edition:
        return 'Winter'
    else:
        return 'na'
df['season'] = df['edition'].apply(extract_season)
```

# 04. Exploring Correlations

## 04.01 Correlation Matrix

```python
# Create a correlation matrix using pandas

df.corr()
```

```python
# Create a subset excluding not useful columns 

sub = df[['athlete_id', 'result_id', 'isTeamSport', 'height', 'weight', 'age']]
```

```python
# Call subset

sub
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Specify desired colormap using the 'cmap' parameter
custom_cmap = "YlGnBu"

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax, cmap=custom_cmap) 
```

### With the data that is possible to correlate, is difficult to tackle the exploratory questions that I had at the beginning. However, based on the analysis, I can say that there's a clear positive correlation between weight and height of participants and between the height of the athletes and team sports.

### Height vs. Weight by Gender

```python
# Try to find a correlation between weight/ height and gender

# Create a data frame of female athletes

female = df[df.gender == 'Female']
```

```python
# Call female

female.head()
```

```python
# Create a data frame of male athletes

male = df[df.gender == 'Male']
```

```python
# Call male

male.head()
```

```python
# Create scatterplot

plt.figure(figsize = (10,5))
plt.scatter(male.height, male.weight, label = "Male", color = 'green')
plt.scatter(female.height, female.weight, label = "Female", color = "purple", alpha= 0.4)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title(" Relationship Between Height and Weight by Gender")
plt.legend()
plt.show()
```

```python
# Create scatterplot Height vs Age

plt.figure(figsize = (10,5))
plt.scatter(male.height, male.age, label = "Male", color = 'green')
plt.scatter(female.height, female.age, label = "Female", color = "purple", alpha= 0.4)
plt.xlabel("Height")
plt.ylabel("Age")
plt.title(" Relationship Between Height and Age by Gender")
plt.legend()
plt.show()
```

```python
# Create scatterplot Weight vs Age

plt.figure(figsize = (10,5))
plt.scatter(male.weight, male.age, label = "Male", color = 'green')
plt.scatter(female.weight, female.age, label = "Female", color = "purple", alpha= 0.4)
plt.xlabel("Weight")
plt.ylabel("Age")
plt.title(" Relationship Between Weight and Age by Gender")
plt.legend()
plt.show()
```

## 04.04 Categorical plot

```python
# Use a histogram to visualize the distribution of the variable 'age'. 


sns.histplot(df['age'], bins = 20, kde = True)
```

```python
# Check again freqneucy distribution of column 'age'

df['age'].value_counts(dropna = False)
```

```python
# Determine sensible categories for the age ranges. 

df.loc[df['age'] < 20, 'age_category'] = 'Youth'
```

```python
df.loc[(df['age'] >= 20) & (df['age'] < 40), 'age_category'] = 'Adult'
```

```python
df.loc[df['age'] >= 40, 'age_category'] = 'Veteran'
```
