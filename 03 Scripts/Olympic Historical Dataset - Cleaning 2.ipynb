# 05. Filling 'na' values on columns 'height' and 'weight'

There are multiple disciplines (sports) in the Olympics. They all have different weight requirements. Therefore it would not be ideal to just impute missing values in columns 'Height' and 'Weight' with the total average. Therefore, is important to at least match it per sport/sport events. The code syntax was adapted from the following Notebook in Kaggle.com: https://www.kaggle.com/code/orcunizmirli/data-analysis-and-data-visualization-for-beginners

### First it is necessary to find unique sports events

```python
# Finding Unique Sport Events

unique_events = pd.unique(df_merged.event_title)
print("Unique Event Count: ", len(unique_events))
unique_events[:10]
```

### Now its time to find Unique Sports

```python
# Finding Unique sports

unique_sports = pd.unique(df_merged.sport)
print("Unique Sport Count: ", len(unique_sports))
unique_sports[:10]
```

```python
# Filling the 'na' values in 'weight' and 'height' 

for each in unique_events:
    
    event_filter = df_merged.event_title == each
    
    weight_mean = df_merged.loc[event_filter, "weight"].mean()
    df_merged.loc[event_filter, "weight"] = df_merged.loc[event_filter , "weight"].fillna(weight_mean)
    
    height_mean = df_merged.loc[event_filter, "height"].mean()
    df_merged.loc[event_filter, "height"] = df_merged.loc[event_filter , "height"].fillna(height_mean)
    
    
# if all event values are NaN 

for each in unique_sports:
    
    sport_filter = df_merged.sport == each
    
    weight_mean = df_merged.loc[sport_filter, "weight"].mean()
    df_merged.loc[sport_filter, "weight"] = df_merged.loc[sport_filter, "weight"].fillna(weight_mean)
    
    height_mean = df_merged.loc[sport_filter, "height"].mean()
    df_merged.loc[sport_filter, "height"] = df_merged.loc[sport_filter, "height"].fillna(height_mean)
```

```python
# Import datetime

from datetime import datetime
```

```python
def isdateYYYYMMDD(date):
    try:
        datetime.strptime(str(date),'%Y-%m-%d') # attempts to convert the given date (which should be in the 'YYYY-MM-DD' format) into a proper date and time object using the specified format.
        return True
    except ValueError:
        return False
    
def feature_age_competing(row):
    if isdateYYYYMMDD(row['born']) and isdateYYYYMMDD(row['start_date']):
        delta_time = datetime.strptime(str(row['start_date']),'%Y-%m-%d') - datetime.strptime(str(row['born']),'%Y-%m-%d')
        return delta_time.days / 365.25
    else:
        return 'na'

    
df_merged['age'] = df_merged.apply(lambda row: feature_age_competing(row), axis=1)
```
